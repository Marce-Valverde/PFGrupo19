Así podemos observar las siguientes tablas:

```python
import pandas as pd
hbo = pd.read_csv('HBO_Max/titles.csv') #Lectura de Archivos
hbo2 = pd.read_csv('HBO_Max/credits.csv') #Lectura de Archivos
```

Encontramos una tabla de mayor dimensión bajo el dataframe 'hbo2':

```python
print('La extensión de la tabla HBO es:', hbo.shape)
print('La extensión de la tabla HBO2 es:', hbo2.shape)
```

De su análisis se observa que la extensión se debe a que cada persona, actor o director, está cargado por cada fila (bajo la columna 'name')

Y como la intención es separar el "cast" conjunto de actores y director por otro lado, se va a utilizar la columna 'role' para trasladar a las columnas que se crearán. Así:

```python
#Podemos ver los valores únicos de esta columna:
print(hbo2['role'].unique())
```

Para ello, se va a ubicar las filas donde el rol sea el de actor, y se llevará a la columna 'cast' que se creará, asignando los valores de la columna 'name'.

```python
#Creamos columna cast y copiamos los valores de las filas name donde la columna rol es igual a ACTOR.
hbo2.loc[hbo2['role'] == 'ACTOR', 'cast'] = hbo2['name']
```

Podemos verificarlo:

```python
hbo2.sample(8)
```

Se va a realizar lo mismo creando la columna director y trasladando sus nombres.

```python
hbo2.loc[hbo2['role'] == 'DIRECTOR', 'director'] = hbo2['name']
```

```python
hbo2.sample(8)
```

Así observando los datos de la tabla:


```python
hbo2.info()
```

No es posible sacar muchas conclusiones dado que reordenando los nombres dentro de 'cast' y 'director' quedará una dimensión mucho menor.

Por existir valores nulos, y se quiere unir los registros de actores, se trabajó a través de Numpy para manejar el método 'join'

```python
import numpy as np

#Reemplazamos los valores nulos por nulos en Numpy

hbo2['cast'] = hbo2['cast'].replace('nan', np.nan)

# Unir los actores en una misma fila, para eso agrupamos por 'id' y a la columna cast agrupe y aplique la función lamba 
#donde a cada valor lo una a través del 'join' y borre previamente los valores nulos. Se reestablece el index.
#para restablecer el índice del resultado.

hbo2_result = hbo2.groupby('id')['cast'].apply(lambda x: ','.join(x.dropna())).reset_index()
```

Se visualiza los resultados:

```python
hbo2_result.sample(8)
```

```python
#Mismo análisis con la columna director.

hbo2['director'] = hbo2['director'].replace('nan', np.nan)
hbo2_result2 = hbo2.groupby('id')['director'].apply(lambda x: ','.join(x.dropna())).reset_index()
```

```python
hbo2_result2.sample(8)
```

El paso siguiente es unir ambas columnas a la tabla hbo2

```python
hbo_unido_con_cast = pd.merge(hbo2, hbo2_result, on='id')
hbo_unido_completo = pd.merge(hbo_unido_con_cast, hbo2_result2, on='id')
hbo_unido_completo.head(8)
```

Para seguir trabajando se realizará una copia de dicha tabla dado que los datos que se necesitan son: id, cast_y y director_y.

```python
hbo_filtrado = hbo_unido_completo[['id', 'cast_y', 'director_y']].copy()
```

Se visualiza el resultado:

```python
hbo_filtrado.head(8)
```

Ahora por último paso se eliminan las filas duplicadas.

```python
hbo_filtrado.drop_duplicates(inplace=True)
```

```python
hbo_filtrado.head(10)
```

Ahora unimos las tablas 'hbo' y la 'hbo2' que fue procesada (ahora es hbo_filtrado) a través de 'id'.

```python
hbo_unido = pd.merge(hbo, hbo_filtrado, on='id')
```

Visualizamos los valores obtenidos

```python
hbo_unido.head(6)
```

Ahora con estos datos reordanamos las columnas y renombramos algunas. Se agregan la columnas 'date_added', 'rating' y 'platform' completando este último con la cadena 'HBO'.

```python
#Se crea una copia con las columnas necesarias, las otras no.
hbo_final = hbo_unido[['id', 'type', 'title', 'director_y', 'cast_y', 'production_countries', 'release_year', 'genres', 'description', 'runtime', 'seasons']].copy()
```

```python
#Se crea las columnas.
hbo_final['date_added'] = ''
hbo_final['rating'] = ''
hbo_final['platform'] = 'HBO'
```

```python
#Se renombra las columnas.
hbo_final.rename(columns={'id': 'show_id', 'director_y': 'director', 'cast_y': 'cast', 'production_countries': 'country', 'genres': 'listed_in', 'runtime': 'duration (mins)'}, inplace=True)
```

Se reordena la tabla.

```python
nuevo_orden = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'listed_in', 'duration (mins)','description', 'platform', 'seasons']

hbo_final = hbo_final.reindex(columns=nuevo_orden)
```

Visualizamos y vemos la descripción de los datos.

```python
print(hbo_final.sample(5))
hbo_final.info()
```

```python
print('La dimensión final de la tabla quedó en: ', hbo_final.shape)
```
