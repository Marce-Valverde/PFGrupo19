```python
netflix['duration'].unique()
```

```python
netflix['duration (min)'] = netflix['duration'].str.extract(r'(\d+) min?')
netflix['seasons'] = netflix['duration'].str.extract(r'(\d+)[^\d]*(?:Season|Seasons)')
```

```python
netflix.drop('duration', axis=1, inplace=True)
```

```python
prime['duration (min)'] = prime['duration'].str.extract(r'(\d+) min?')
prime['seasons'] = prime['duration'].str.extract(r'(\d+)[^\d]*(?:Season|Seasons)')
```

```python
disney['duration (min)'] = disney['duration'].str.extract(r'(\d+) min?')
disney['seasons'] = disney['duration'].str.extract(r'(\d+)[^\d]*(?:Season|Seasons)')
```

 Del dataframe principal elimina corchetes y comillas de la columna 'country'.

```python
# Crea una tabla de traducción para eliminar los caracteres específicos
translation_table = str.maketrans("", "", "[]'")

# Elimina corchetes y comillas de la columna utilizando la tabla de traducción
df['country'] = df['country'].str.translate(translation_table)

df.sample(15)
```

Evolución del contenido de Películas y Series por Año.

```python
df_peliculas = df[df['type'] == 'Movie']

conteo_por_anio = df_peliculas['release_year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(conteo_por_anio.index, conteo_por_anio.values, marker='o', linestyle='-', color='b')
plt.xlabel('Año')
plt.ylabel('Cantidad de Películas')
plt.title('Evolución de Películas por Año')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
df_series = df[df['type'] == 'TV Show']

# Contar la cantidad de películas por año
conteo_por_anio = df_series['release_year'].value_counts().sort_index()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(conteo_por_anio.index, conteo_por_anio.values, marker='o', linestyle='-', color='b')
plt.xlabel('Año')
plt.ylabel('Cantidad de Series')
plt.title('Evolución de Series por Año')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

Evolución del contenido por Año y Plataforma.

```python
plataformas = ['disney', 'hbo', 'netflix', 'prime']
df_plataformas = df[df['platform'].isin(plataformas)]

conteo_por_anio_plataforma = df_plataformas.groupby(['release_year', 'platform']).size().unstack()

plt.figure(figsize=(10, 6))

for plataforma in conteo_por_anio_plataforma.columns:
    plt.plot(conteo_por_anio_plataforma.index, conteo_por_anio_plataforma[plataforma], marker='o', linestyle='-', label=plataforma)

plt.xlabel('Año')
plt.ylabel('Cantidad de contenido')
plt.title('Evolución del contenido por Año y Plataforma')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```
