 Del dataframe principal elimina corchetes y comillas de la columna 'country'.

```python
# Crea una tabla de traducción para eliminar los caracteres específicos
translation_table = str.maketrans("", "", "[]'")

# Elimina corchetes y comillas de la columna utilizando la tabla de traducción
df['country'] = df['country'].str.translate(translation_table)

df.sample(15)
```

Se trabaja con dataframes auxiliares para unificar las descripciones de los paises.

```python
df2 = pd.DataFrame()
df2['country'] = df['country']   #me llevo la col. country del df ppal.
df2
```

```python
df2 = pd.DataFrame()
df3 = pd.merge(df2,paises,how='outer',left_index=True, right_index=True) # merge con paises
df3
```

Reemplazo los valores nulos de country antes de hacer el merge.

```python
df['country'].fillna('unknown', inplace=True) 
df.sample(15)
```

Realiza un merge para unir con df ppal. de descripción de paises.

```python
df = pd.merge(df, df3, left_on='country', right_on='country', how='left')
df
```
