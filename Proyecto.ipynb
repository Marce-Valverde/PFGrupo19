Reordenamos las columnas de HBO.

```python
nuevo_orden = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration (min)', 'listed_in', 'description']
```

```python
hbo = hbo.reindex(columns=nuevo_orden)
hbo
```

Se crea otra columna para identificar cada plataforma.

```python
netflix['platform']='netflix'
disney['platform']='disney'
prime['platform']='prime'
hbo['platform']='hbo'
```

Podemos ver que los dataframes de Netflix, Hulu y Disney tienen en su columna "duration" la duracion como las temporadas (Season...) por lo que se va a retrabajar estos df separando agregando la columna seasons.

```python
netflix['duration'].unique()
```

```python
netflix['duration (min)'] = netflix['duration'].str.extract(r'(\d+) min?')
netflix['seasons'] = netflix['duration'].str.extract(r'(\d+)[^\d]*(?:Season|Seasons)')
```

```python
netflix.drop('duration', axis=1, inplace=True)
```

```python
prime['duration (min)'] = prime['duration'].str.extract(r'(\d+) min?')
prime['seasons'] = prime['duration'].str.extract(r'(\d+)[^\d]*(?:Season|Seasons)')
```

```python
prime.drop('duration', axis=1, inplace=True)
```

```python
disney['duration (min)'] = disney['duration'].str.extract(r'(\d+) min?')
disney['seasons'] = disney['duration'].str.extract(r'(\d+)[^\d]*(?:Season|Seasons)')
```

```python
disney.drop('duration', axis=1, inplace=True)
```
